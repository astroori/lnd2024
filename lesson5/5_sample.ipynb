Для успешного решения задачи данного урока вам может понадобиться визуализировать каждый элемент данных по-отдельности. В то же время для большей наглядности представления данных можно визуализировать каждый из 4-х параметров из файла на отдельном графике, а все графики вывести одновременно.

```python
import matplotlib.pyplot as plt

f = open('5_1.txt', 'r')
data = []
for line in f:
    p1, p2, p3, p4 = map(int, line.split())
    data.append([p1, p2, p3, p4])
f.close()

plt.figure(figsize=(15, 5))
for i, x in enumerate(data):
    for j in range(4):
        plt.subplot(1, 4, j + 1)
        plt.scatter(x[j], i, c = 'blue')
plt.show()
```

Внимательно изучите данный пример. Здесь на каждом из графиков изображён один из параметров. По горизонтальной оси отмечено значение параметра, по вертикальной оси - номер соответствующего объекта в файле (от 0 до 125). Конструкция plt.subplot позволяет указать номер графика на котором будет визуализироваться требуемая точка.
