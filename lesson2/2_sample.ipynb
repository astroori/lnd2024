В случае с данной задачей в каждой строке файла находится два числа - данные о росте и весе, поэтому считывать и сохранять данные из файла можно с использованием следующего исходного кода:

```python
f = open('2_1.txt', 'r')
data = []
for line in f:
    h, w = map(int, line.split())
    data.append([h, w])
```

В результате в спике data будут расположены пары чисел из каждой строки файла, при этом сам список data будет двумерным:

```python
print(data)
```
