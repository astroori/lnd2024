В файле 7_1.txt содержится информация об одобрении кредитных заявок клиентам банка в следующем формате. Первый параметр - возраст клиента, второй параметр - среднемесячная заработная плата за последний год, третий параметр - информация о наличии или отсутствии в собственности недвижимости (0 - недвижимость отсутствует, 1 - недвижимость имеется), четвёртый параметр - информация об одобрении и неодобрении кредита (0 - в кредите отказано, 1 - кредит одобрен). По введённым данным потенциального клиента банка (возрасту, зарплате и наличию недвижимости) на основании представленных данных попытайтесь предсказать:
Задача 1: одобрят или не одобрят данному клиенту кредит?
Задача 2: в скольки банках из 10 данный клиент вероятнее всего получит одобрение кредита?
