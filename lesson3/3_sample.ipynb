При считывания файла в данном уроке необходимо обратить внимание на то, что данные в файле имеют двумерный характер - всего 100 объектов, для каждого объекта 2 параметра. Поэтому считывание лучше организовать следующим образом:

```python
f = open('3_1.txt', 'r')
data = []
for line in f:
    p1, p2 = map(int, line.split())
    data.append([p1, p2])
f.close()
print(data)
```
